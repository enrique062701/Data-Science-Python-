```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m_t = np.array([20,50,70,100,150,200])
m_t = m_t/1000
halfperiod = np.array([33.0,20.5,17.5,15.5,13.5,12.0])

f = 1/((2*halfperiod)/1000)
fsquared = f**2

fsquared = np.insert(fsquared, 0, 0)
m_t = np.insert(m_t, 0, 0)

print (fsquared)
```

```python
results, cov = np.polyfit(m_t, fsquared, 1, cov=True)
bestfit = m_t*results[0] + results[1]

plt.plot(m_t, bestfit, color="red", label="Line of best fit")
plt.scatter(m_t, fsquared, label="Data")
plt.xlabel("Mass Applying Tension On String (kg)")
plt.ylabel("Frequency of Oscillation Squared (Hz^2)")
plt.title("Best fit from data demonstrating the effect of varying tension\n on the fundamental frequency of an oscillating string")
plt.legend()
print ("f^2 =", results[0],"+",results[1])
```

```python
results, cov = np.polyfit(m_t, fsquared, 1, cov=True)
bestfit = m_t*results[0] + results[1]

print(cov[0,0], np.sqrt(cov[0,0]))

bestfitmax = m_t*(results[0]+np.sqrt(cov[0,0])) + results[1]
print (results[0]+np.sqrt(cov[0,0]),"+",results[1]) ##max equation
bestfitmin = m_t*(results[0]-np.sqrt(cov[0,0])) + results[1]
print (results[0]-np.sqrt(cov[0,0]),"+",results[1]) ##max equation


bestfit_b0 = m_t*results[0]
plt.fill_between(m_t, bestfitmin,bestfitmax,color='red',alpha=.2, label="Uncertainty range in LoBF")

plt.scatter(m_t, fsquared, label="Data")
##plt.plot(m_t, bestfit, color="red")
plt.plot(m_t, bestfitmax, color="red")
plt.plot(m_t, bestfitmin, color="red")

plt.xlabel("Mass Applying Tension On String (kg)")
plt.ylabel("Frequency of Oscillation Squared (Hz)")
plt.title("Best fit and uncertainties from data demonstrating the effect of varying\n tension on the fundamental frequency of an oscillating string")
plt.legend()
```
